```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

def extract_text(url):
    r = requests.get(url)
    soup = BeautifulSoup(r.text,"html.parser")
    para = soup.find_all('p')
    text = ' '.join([p.get_text() for p in para])
    return text

url = "https://www.ndtv.com/india#pfrom=home-ndtv_mainnavgation"
r = requests.get(url)
print(r)
soup = BeautifulSoup(r.text, "html.parser")



main_body=[]
urls=[]
titles = []
all_page_urls=[]
while url:
    # Fetch the current page
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')

    # Find the next page link
    next_page_link = soup.find('a', class_="btnLnk arrowBtn next")

    # If there's a next page link, extract the URL and continue, otherwise break the loop
    if next_page_link:
        next_page_url = next_page_link.get('href')
        all_page_urls.append(url)
        url = next_page_url
    else:
        all_page_urls.append(url)
        break

for page_url in all_page_urls:
    response = requests.get(page_url)
    soup = BeautifulSoup(response.text, 'html.parser')
    title = soup.find_all("h2",class_='newsHdng')
    for i in title:
        titles.append(i.text)
    h2tags = soup.find_all('h2')
    for tags in h2tags:
        links=tags.find_all('a')
        for link in links:
            link_url = link.get('href')
            urls.append(link_url)
            body = extract_text(link_url)
            main_body.append(body)

df=pd.DataFrame({'URL':urls,
                 'TITLE': titles,
                  'MAIN BODY' : main_body
                 })
print(df)


df.to_csv('/content/drive/MyDrive/finalnews.csv', index=False)
print("Data is saved")
# link = soup.find_all('div',class_="container__field-links container_vertical-strip__field-links")
#print(link)
# Extract the URLs from the link elements and store them in a list
# links = []
# link_a = link[1].find_all('a')  # The second 'container_
# link = link.get('href')
# for link_element in link_a:
#     links.append(link_element['href'])
# print(links)
# link= soup.find("a",class_='href')
# print(link)
# print(soup)
# headers = soup.find_all("font",style_="vertical-align: inherit;")
# print(headers)


```
